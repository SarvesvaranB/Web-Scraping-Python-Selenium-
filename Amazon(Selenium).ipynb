```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
driver=webdriver.Chrome(r'E:/chromedriver')
driver.get("https://www.amazon.in/")
driver.maximize_window()
```

```python
item=driver.find_element_by_id('twotabsearchtextbox')
item.send_keys('Samsung Phones')
search=driver.find_element_by_id('nav-search-submit-button')
search.click()
```

```python
name=[]
try:
    pname=driver.find_elements_by_xpath('//span[@class="a-size-medium a-color-base a-text-normal"]')
    for phone in pname:
         name.append(phone.text)
except:
        name.append('-')
print(name)
```

```python
price=[]
try:
    pprice=driver.find_elements_by_xpath('//span[@class="a-price-whole"]')
    for p in pprice:
        price.append(p.text)
except:
    price.append('-')
print(price)
```

```python
import pandas as pd
finallist=zip(name,price)
df=pd.DataFrame(finallist)
df.columns=['Mobile_Name','Mobile_Price']
```
